## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
max_h = np.array([[12, 11], [7, 10]])
max_j = np.array([[0, 1, 2],[7, 10, 4],[3, 2, 0]])
print(np.linalg.det(max_h))
print(np.linalg.det(max_j))
```

## 3.3 Find $H^{-1}$ and then find $J^{-1}$

```
max_h_inverse = np.linalg.inv(max_h)
print(max_h_inverse)
max_j_inverse = np.linalg.inv(max_j)
print(max_j_inverse)
```
